```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
data = pd.read_csv('DATA-2023/anandVihar-delhi.csv', index_col=0)
```

```python
file_names=os.listdir('DATA-2023')
```

```python
print(file_names[0].split('-')[0])
data['Location'] = file_names[0].split('-')[0]
```

```python
data.columns
```

```python
data.drop(columns=['Benzene (µg/m³)', 'Toluene (µg/m³)', 'Xylene (µg/m³)',
       'O Xylene (µg/m³)', 'Eth-Benzene (µg/m³)', 'MP-Xylene (µg/m³)',
       'AT (°C)', 'RH (%)', 'WS (m/s)', 'WD (deg)', 'RF (mm)', 'TOT-RF (mm)',
       'SR (W/mt2)', 'BP (mmHg)', 'VWS (m/s)'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data['Timestamp'].dtype
```

```python
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
```

```python
data['Timestamp'].dtype
```

```python
data['Month'] = data['Timestamp'].dt.month
data['Day'] = data['Timestamp'].dt.day
```

```python
data.head()
```

```python
y = data.groupby('Month')['PM2.5 (µg/m³)'].mean()
y2 = data.groupby('Month')['PM10 (µg/m³)'].mean()
```

```python
y
```

```python
plt.figure(figsize=(10,5))
x = np.arange(1,13)
bar_width = 0.4
plt.bar(x-bar_width/2,y,label='PM 2.5',width=bar_width)
plt.bar(x+bar_width/2,y2,label='PM 10',width=bar_width)
plt.xlabel('Month')
plt.ylabel('avg PM2.5 level')
plt.title('Month Wise Average PM2.5 Level In Air')
plt.tight_layout()
plt.legend(loc = 'upper right')
plt.xticks(x)
plt.show()
```
